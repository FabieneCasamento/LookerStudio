REGEXMATCH

Identifica se parte do texto corresponde a uma expressão regular.

Exemplo de uso
REGEXMATCH("Planilhas"; "P.a")

Sintaxe
REGEXMATCH(texto; expressão_regular)

texto: o texto a ser testado em relação à expressão regular.

expressão_regular: a expressão regular para testar o texto.

https://support.google.com/docs/answer/3098292?hl=pt-BR
